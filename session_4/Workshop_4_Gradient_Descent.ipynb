## 2. Gradient Descent

$$
\begin{aligned}
\theta_0 &:=\theta_0-\alpha \frac{1}{m} \sum_{i=1}^m\left[\left(h_\theta\left(x_i\right)-y_i\right) x_i^{(0)}\right] \\
\theta_1 &:=\theta_1-\alpha \frac{1}{m} \sum_{i=1}^m\left[\left(h_{\theta}\left(x_i\right)-y_i\right) x_i^{(1)}\right]
\end{aligned}
$$
## x1 is equal to 1

```python
slope = 1
bias = 1
alpha = 0.0001

for iteration in range(100):

    #gradient descent for slope
    slope_sample_gradient_list = []
    for i in range (len(X)):
        h = slope*X[i]+bias
        slope_one_sample_gradient = ((h-y[i])*X[i])
        slope_sample_gradient_list.append(slope_one_sample_gradient)
    slope -=alpha*sum(slope_sample_gradient_list)
    
    
    bias_sample_gradient_list = []
    for i in range(len(X)):
        h = slope*X[i]+bias
        bias_one_sample_gradient = (h-y[i])
        bias_sample_gradient_list.append(bias_one_sample_gradient)
    bias -=alpha*sum(bias_sample_gradient_list)


plt.figure(figsize=[4,3])
plt.xlim(0,15)
plt.ylim(0,30)

y_pred = []
for i in X:
    y_pred.append(i*slope+bias)

plt.scatter(X,y)
plt.plot(X,y_pred,c="red")
```
